## Exercise 1: Putting it all together

Write a program that does the following:

- Prints the numbers 1 to 100
- For the  numbers 10 to 25 (including 10 and 25), instead of printing the number, print the word "cheese".
- For numbers 55 to 100 (including 55, but not including 100), instead of printing the number, print the word "cake".
- For 100 print the word "Done!".

**Important Python concepts:** print(), if, elif, else, for

````python
# Write code here
#Could be simplified to i >= 10 and i <= 25
for i in range(1,101,1):
    if i < 10:
        print('{}'.format(i))
    elif i <= 25:
        print('cheese{}'.format(i))
    elif i < 55:
        print('{}'.format(i))
    elif i <= 99:
        print('cake{}'.format(i))
    elif i == 100:
        print('Done!')
    else:
        print('Outside intended range')
````

## Exercise 2

Write a program which repeatedly reads numbers until the user enters â€œdoneâ€. Once â€œdoneâ€ is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

What the code should look like when running:
```
Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.33333333333
```

**Important Python concepts:** print(), try, except, break, while

````python
total = 0
count = 0
average = 0
while True:
    try:
        number = input('Enter a number: ')
        number = float(number)
        total += number
        count = count + 1
        average = total / count
    except:
        if number != 'done':
            print('Invalid Input')
        else:
            break
print('Sum: {}, Count: {}, Average: {}'.format(total,count,average))
````
