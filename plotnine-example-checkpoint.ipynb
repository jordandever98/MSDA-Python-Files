In the next example, we will first create count of class and then plot it

```python
import pandas as pd
```

```python


mpg_counts = mpg['class'].value_counts().reset_index() #reset index to not let pandas auto decide 
mpg_counts.columns = ['class', 'count'] # df w/ 7 rows x columns

# Get a sorted list of vehicle classes based on count
ordered_classes = mpg_counts.sort_values(by='count', ascending=False)['class'].tolist()

# Convert 'class' to a categorical variable with the defined order
mpg_counts['class'] = pd.Categorical(mpg_counts['class'], categories=ordered_classes, ordered=True)
```

```python
(ggplot(mpg_counts, aes(x='class', y='count', label='count'))
        + geom_col()
        + geom_text(aes(y = 'count + 2'))
        + labs(x="Vehicle Class", y="Count")
        + theme_minimal())
```

Above I used y = 'count + 2' inside aes() and geom_text(). You can use va = "botom" inside geom_text() instead and it will nicely place the labels just above the bars. 

```python
mpg_counts2 = mpg['class'].value_counts().reset_index() 
mpg_counts2.columns = ['class', 'count'] #creating new DF called mpg_count2 that is unordered

# to order the data need to recreate an ordered variable. 
# Taking the obj, then sort values, ascending = F (Desc), then arrange based on ['count']
ordered_classes = mpg_counts.sort_values(by='count', ascending=False)['class'].tolist()


# then need to do factorization on the data.
# Convert 'class' to a categorical variable with the defined order
mpg_counts['class'] = pd.Categorical(mpg_counts['class'], categories=ordered_classes, ordered=True)
```

```python
(ggplot(mpg_counts2, aes(x='class', y='count', label='count'))
        + geom_col()
        + geom_text(va='bottom') #dont have vjust but can use bottom to place the numbers on the bars
        + labs(x="Vehicle Class", y="Count")
        + theme_minimal())

#unordered visiualization, ordered by vehicle class name
```
